```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("exams.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[['math score', 'reading score', 'writing score']].mean()
```

```python
df[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')

plt.title("Average Scores by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.show()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')

plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')

plt.title("Impact of Test Preparation on Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')

plt.title("Average Scores by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', figsize=(10,5))

plt.title("Average Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

corr = df[['math score', 'reading score', 'writing score']].corr()

plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```
